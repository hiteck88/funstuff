### Scatter plot 

```python
gapminder = pd.read_csv('gapminder.csv')
gapminder.head()
```

```python
plt.scatter(gapminder.life_exp, gapminder.gdp_cap)
#plt.xscale('log')
plt.show()
```

#### modify the size of dots

```python
plt.scatter(gapminder.gdp_cap, gapminder.life_exp, s = 2)  # size 
plt.show()
```

```python
# try to make the dots even larger, according to the population 

np_pop = np.array(gapminder.population)/1000000
plt.scatter(gapminder.gdp_cap, gapminder.life_exp, s = np_pop, alpha = 0.5)  # size and transparency

# Additional customizations
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')

plt.show()
```

#### further customisation

```python
plt.scatter(gapminder.population, gapminder.life_exp)
# Add axis labels
plt.xlabel('population')
plt.ylabel('life expectancy')


# Add title
plt.title('World development')

# change the lables for the axis using ticks
# Definition of tick_val and tick_lab
tick_val = [0.2*10**9, 0.6*10**9, 1.0*10**9]
tick_lab = ['2b', '6b', '10b']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)


# After customizing, display the plot
plt.show()
```

### Histograms

```python
# plot the life expectancy
plt.hist(gapminder.life_exp, bins = 20)  # default bin = 10 (bin is how many slots, fewer, the thicker each one.)
plt.show()
```
